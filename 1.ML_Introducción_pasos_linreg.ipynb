## Introducción al machine learning

Hay diferentes tipos de machine learning, 

### Aprendizaje supervisado / no supervisado:

Hay cuatro grandes categorias: aprendizaje supervizado, aprendizaje no supervizado, semi supervizado, aprendizaje reforzado.

1. Aprendizaje supervisado: En este los datos de entrenamiento dados al algoritmo incluyen las soluciones deseadas, por ejemplo, clasificación, regresión, KNN, SVM, árboles de decisión y random forests, redes neuronales.

1. Aprendizaje no supervizado: Los datos de entrenamiento no tienen etiqueta, por ejemplo, clustering, detección de anomalias, visualización y reducción de dimensionalidad.

1. Aprendizaje semi supervizado: Cuando se tienen datos parcialmente etiquetados.

1. Aprendizaje por refuerzo: El sistema de aprendizaje, lllamado agente, observa el ambiente, selecciona y ejecuta acciones y obtiene recompensas o penalidades, debe aprender por si solo cual es la mejor estrategia (politica) para tener la mayor cantidad de recompensas con el tiempo, una politica define lo que el agente debe hacer dada una situación.

### Batch y aprendizaje en linea.

Este depende si nuestro sistema puede o no aprender de forma incremental a partir de un flujo de datos entrantes.

1. Aprendizaje batch: Para este el sistema no puede aprender incremental, tiene que ser entrenado en datos disponibles, el sistema es entrenado, lanzado a producción y corre sin aprender más.

1. Aprendizaje en linea: El sistema aprende incremental, aprende de nuevos datos al momento, lo rápido que se adapta el modelo a datos cambiantes se llama tasa de aprendizaje, una alta tasa de aprendizaje puede llevar a que el sistema olvide datos viejos muy rápido, una tasa baja va a hacer que el sistema sea menos sensible a ruido en los nuevos datos u outliers.

### Aprendizaje basado en instancias versus aprendizaje basado en modelos

1. Basado en instancias: El sistema aprende por ejemplos, luego generaliza a nuevos casos usando una medida de similitud comparado al ejemplo dado.

1. Basado en modelos: Se construye un modelo de estos ejemplos y usamos ese modelo para hacer predicciones.

### Dificultades del machine learning

1. Cantidad insuficiente de datos de entrenamiento: Hay papers que muestran que con la cantidad suficiente de datos los algortimos tienden a mejorar, igualmente no hay que dejar de preocuparse por cual algoritmo usar.

1. Datos de entrenamiento no representativos: Si los datos no representan el total de la población se van a tener modelos con resultados no significativos, si la muestra es pequeña se puede tener ruido de muestreo (datos no representativos por el azar), si el metodo de muestreo es malo podemos igual tener muestras no representativas en muestras grandes, se le llama sesgo de muestreo.

1. Datos de poca calidad: si hay muhcos otliers o muchos datos perdidos, aquí entra la ingenieria de variables.

1. Variables irrelevantes: Basicamente hacer ingenieria de variables, seleccionar las útiles, reducir dimensionalidad, nuevas variables, etc

1. Sobreajustar los datos de entrenamiento: El modelo tiene muy buen rendimiento en los datos de entrenamiento pero no generaliza a nuevos datos, para esto se puede escoger un modelo más simple (regularizar), tener mas datos de entrenamiento o reducir el ruido en los datos de entrenamiento (outliers, errores, perdidos).

Un hiperparámetro es un parámetro de aprendizaje del algoritmo (no del modelo).

1. Subajustar los datos de entrenamiento: El opuesto del anterior, se puede mejorar escogiendo un modelo más complejo, encontrando mejores variables para el algoritmo o reducir las restricciones en el modelo.

### Probar y validar

Se dividen nuestros datos en entrenamiento y prueba, la proporcion de error en nuevos casos se llama el error de generalización y se obtiene un estimado evaluando nuestro modelo en los datos de prueba.

#### Ajuste de hiperparámetros y selección del modelo

Si se ajustan los hiperparámetros con los datos de prueba puede que el modelo no generalice luego, para esto se usa unos datos de validación que son parte de los datos de entrenamiento, se entrena el modelo con los datos de entrenamiento menos el de validación y se escoge el que mejor rinda, con sus hiperparámetros, con los datos de validación y se evalua en los datos de prueba.

Se debe usar validación cruzada, saundo mucho datos de validación pequeños, cada modelo es evaluado una vez por conjunto de validación, promedia las evaluaciones y se obtiene una medida del rendimiento mucho mejor.

También se puede tener otros conjunto de datos que se aparta del conjunto de entrenamiento (train-dev set), luego que el modelo es entrenado en los datos de entrenamiento se puede evaluar en el train-dev set, si rinde bien el modelo no sobreajusta los datos de entrenamiento, si rinde mal en el conjunto de validación el problema debe venir de discrepancia de datos.

### No free lunch

Para distintos modelos hay diferentes supuestos a tener en cuenta, David Wolpert demonstró que si no se hacen supuestos de los datos no hay razón para preferir un modelo sobre ningún otro

#### Seleccionar una medida de rendimiento

Con que medida vamos a calificar el rendimiento del modelo, en el caso de la regresión usamos el RMSE,

$$
RMSE(X, h) = \sqrt{\frac{1}{m}\sum (h(x^{(i)})-y^{(i)})^2}
$$

- m es el numero de observaciones en el dataset.

- $x^{(i)}$ es el vector de los valores de las variables de la observación i, $y^{(i)}$ es la variable respuesta observada.

- $h$ es la función que da la predicción, también llamada la hipótesis. $\hat{y}^{(i)} = h(x^{(i)})$.

- $RMSE(X, h)$ es la función de costo usando la hipótesis $h$.

En algunos casos se prefieren otras medidas, por ejemplo, si hay muchos outliers se puede preferir el MAE, ambas medidas son distancias entre dos vectores, el RMSE es la distancia euclidiana, el MAE es la norma o la norma Manhattan

### Crear un conjunto de prueba

Debemos crear este conjunto lo antes posible sin mirar los datos, si miramos los datos podriamos encontrar algún patrón que nos haga tener 
sesgo al escoger el conjunto de prueba, esto es llamado *data snooping bias*.

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                        bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
```

### Escalado de variables

Es una de las transformaciones más importantes, los algoritmos en su mayoría no rinden bien con variables de diferente escala, los dos escalados más comunes es *min max scaling* y *estandarización*

1. min - max scaling: para que todos los valores queden entre 0 y 1, restamos el menor valor y dividimos por el maximo menos el minimo.

1. estandarización: primero resta por la media y divide por la desviación estándar, estandarización se ve menos afectado por outliers

Importante, como todas las transformaciones, solo ajustamos los escalados a los datos de entrenamiento, no a los datos completos. Sólo después se pueden utilizar para transformar todos los datos.

### Pipelines de transformación



```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

Cuando llamamos fit() en una pipeline, llama fit_transform() en todas las transformaciones, pasando el resultado de cada paso al siguiente hasta que llega al último estimador, para el cual llama fit(), las pipelines se pueden combinar con pandas

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs),
])

housing_prepared = full_pipeline.fit_transform(housing)
```

### Evaluación usando validación cruzada

El codigo divide el conjunto de entrenamiento en 10 subconjuntos llamados *folds*, entrena y evalua el árbol de decision 10 veces, escogiendo un *fold* diferente cada vez y entrenando el modelo en los otros 9, el resultado es un arreglo que contiene el puntaje de 10 evaluaciones.

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring = 'neg_mean_squared_error', cv = 10)
tree_rmse_scores = np.sqrt(-scores)
```

La función de validación cruzada espera una función de utilidad (más es mejor) en vez de una función de costo (menos es mejor), la función de puntaje es lo contrario al MSE, por eso calculamos (-score)

```python
def display_scores(scores):
    print('Scores:', scores)
    print('Media:', scores.mean())
    print('Desviación estándar', scores.std())
```

```python
display_scores(tree_rmse_scores)
```

3. Añadir una transformación a la pipeline para seleccionar los atributos más importantes

```python
from sklearn.base import BaseEstimator, TransformerMixin

def indices_of_top_k(arr, k):
    return np.sort(np.argpartition(np.array(arr), -k)[-k:])

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k
    def fit(self, X, y=None):
        self.feature_indices_ = indices_of_top_k(self.feature_importances, self.k)
        return self
    def transform(self, X):
        return X[:, self.feature_indices_]
```
